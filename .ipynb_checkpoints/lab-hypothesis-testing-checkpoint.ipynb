**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#H0: mu Dragon HP <= Grass HP
#H1: mu Dragon HP > Grass HP

df_dragon = df[df["Type 1"] == "Dragon"]["HP"]
df_grass = df[df["Type 1"] == "Grass"]["HP"]

result = st.ttest_ind(df_dragon, df_grass, equal_var=False, alternative="greater")

t_stat = result.statistic
p_value = result.pvalue

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
from scipy.stats import ttest_ind 

cols = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
alpha = 0.05

for stat in cols:
    stat_legend = df[df["Legendary"] == True][stat]
    stat_non = df[df["Legendary"] == False][stat]
    t_stat, p_value = ttest_ind(stat_legend, stat_non, equal_var=False, alternative="two-sided")

    print(f"\nStat: {stat}")
    print(f"T-statistic: {t_stat:.2f}")
    print(f"P-value: {p_value:.2f}")

    if p_value < alpha:
        print("Reject H0")
    else:
        print("Fail to reject H0")
```

**Challenge 2**

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
school = [-118, 34]
hospital = [-112, 37]

long = df["longitude"]
lat = df["latitude"]

euc_school = []
euc_hospital = []

for x in range(len(long)):
    euc_school.append(((long[x]-school[0])**2 + (lat[x]-school[1])**2)**0.5)
    euc_hospital.append(((long[x]-hospital[0])**2 + (lat[x]-hospital[1])**2)**0.5)

df["euc_school"] = euc_school
df["euc_hospital"] = euc_hospital

df.head
```

```python
close_school = []

for x in range(len(df["euc_school"])):
    if df["euc_school"][x] < 0.5:
        close_school.append(True)
    else:
        close_school.append(False)

close_hospital = []

for x in range(len(df["euc_hospital"])):
    if df["euc_hospital"][x] < 0.5:
        close_hospital.append(True)
    else:
        close_hospital.append(False)


df["close_school"] = close_school
df["close_hospital"] = close_hospital


df["close_any"] = df["close_school"] | df["close_hospital"]
```

```python
group_close = df[df["close_any"] == True]["median_house_value"]
group_far = df[df["close_any"] == False]["median_house_value"]

t_stat, p_value = ttest_ind(group_close, group_far, equal_var=False, alternative="two-sided")

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```
